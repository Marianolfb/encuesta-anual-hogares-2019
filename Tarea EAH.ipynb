```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from funpymodeling import status, profiling_num, freq_tbl
from ydata_profiling import ProfileReport
```

```python
data=pd.read_csv('encuesta-anual-hogares-2019.csv',sep=',')
pd.set_option('display.max_columns', None)
data.head()
```

```python
##data_s=data.sample(frac=0.01)
data.columns
```

```python
data = data.drop({'id','hijos_nacidos_vivos'},axis=1)
```

```python
ingreso_f, saved_bins_f = pd.qcut(data['ingresos_familiares'],q=8,retbins=True)
```

```python
freq_tbl(ingreso_f)
ingreso_f.value_counts()
```

```python
ingreso_perc, ret_bins_perc = pd.qcut(data['ingreso_per_capita_familiar'],q=10,retbins=True)
```

```python
ingreso_perc.value_counts()
```

```python
data['ingreso_total_lab'] = pd.qcut(data['ingreso_total_lab'], q=10, duplicates='drop')
data['ingreso_total_no_lab'] = pd.qcut(data['ingreso_total_no_lab'], q=4, duplicates='drop')
data['ingreso_total_lab'].value_counts()
```

```python
data['ingresos_familiares']=ingreso_f
data['ingreso_per_capita_familiar'] = ingreso_perc
```

```python
data['ingresos_familiares'].value_counts()
```

```python
data.head()
```

```python
data['ingreso_total_no_lab'].value_counts()
```

```python
data['edad'] = pd.cut(data['edad'],bins=5)
```

```python
data['edad'].value_counts()
```

```python
data['comuna'] = data['comuna'].apply(str)
```

```python
data['nhogar'] = data['nhogar'].apply(str)
```

```python
data['años_escolaridad'].unique()
```

```python
data['años_escolaridad'] = data['años_escolaridad'].replace('Ningun año de escolaridad aprobado','0')
```

```python
data['años_escolaridad'].unique()
```

```python
data['años_escolaridad'] = data['años_escolaridad'].astype(float)
```

```python
data['años_escolaridad'] = pd.qcut(data['años_escolaridad'],q=5)
```

```python
data['años_escolaridad'].value_counts()
freq_tbl(data['años_escolaridad'])
```

```python
data = data.dropna(subset=['situacion_conyugal', 'sector_educativo',
       'lugar_nacimiento', 'afiliacion_salud'])
```

```python
data = data.reset_index(drop=True)
```

```python
data['nivel_max_educativo'].fillna('desconocido',inplace=True)
```

```python
data['años_escolaridad'] = data['años_escolaridad'].cat.add_categories('desconocido')
```

```python
data['años_escolaridad'].fillna('desconocido',inplace=True)
```

```python
freq_tbl(data['años_escolaridad'])
```
